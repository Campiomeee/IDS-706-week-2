```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score


cols = ["salary_usd", "years_experience"]
X = df[cols]

# Standardize 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# try out different values of K
k_values = range(2, 11)          
inertias = []
silhouettes = []
models = {}

for k in k_values:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    km.fit(X_scaled)
    models[k] = km
    
    # store the inertia and silhouettes
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_scaled, km.labels_))

# save the results into a dataframe
results = pd.DataFrame({
    "k": list(k_values),
    "inertia": inertias,
    "silhouette": silhouettes
})
print(results)
                       
# Visualize the inertia
plt.plot(results["k"], results["inertia"], marker="o")
plt.xlabel("k"); plt.ylabel("Inertia (WSS)")
plt.title("Elbow Method")
plt.show()

# Visualize the silhouette
plt.plot(results["k"], results["silhouette"], marker="o")
plt.xlabel("k"); plt.ylabel("Silhouette score")
plt.title("Silhouette vs k")
plt.show()

# find the best value of K
best_k = results.loc[results["silhouette"].idxmax(), "k"]
print("Best k by silhouette:", int(best_k))

# fit final model and attach labels
final_km = models[int(best_k)]
df.loc[X.index, "cluster"] = final_km.labels_
```

```python
plt.scatter(X[cols[0]], X[cols[1]], c=df.loc[X.index, "cluster"], cmap="viridis", alpha=0.7)
plt.xlabel(cols[0])
plt.ylabel(cols[1])
plt.title("K-means Clusters")
plt.show()
```

```python
# Try different method, Linear regression.

from sklearn.linear_model import LinearRegression

X = df[["years_experience"]]      # independent variable
y = df["salary_usd"]

linreg = LinearRegression()
linreg.fit(X, y)

print("Intercept:", linreg.intercept_)
print("Slope:", linreg.coef_[0])
print(f"Function: Salary_USD = {linreg.coef_[0]} x Years_Experience + {linreg.intercept_}")

# Predictions
y_pred = linreg.predict(X)

# Plot scatter + regression line
plt.scatter(X, y, alpha=0.6, label="Data")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary_USD")
plt.title("Linear Regression: Salary vs Experience Year")
plt.legend()
plt.show()
```
